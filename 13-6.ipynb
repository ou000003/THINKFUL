```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dt = pd.read_csv('WELLCOME_APCspend2013_forThinkful.csv', encoding = 'unicode_escape')
dt.describe()
```

```python
dt['Journal title'].unique()
```

```python
dt['Journal title lower'] = dt['Journal title'].apply(lambda x: str(x).translate({ord(c): " " for c in "!@#$%^*()[]{};:,./<>?\|`~-=_+"}).lower())
dt['Journal title lower'] = dt['Journal title lower'].apply(lambda x: x.replace("journal of", "").strip())
dt['Journal title lower'] = dt['Journal title lower'].apply(lambda x: x.replace("&", "and").strip())
dt.describe()
```

```python
dt['Journal title lower'].unique()
```

```python
print(dt.isna().sum())
dt[dt['Journal title'].isna()== True]
```

```python
dt.groupby('Journal title lower')['Article title'].count().sort_values(ascending = False).head()
```

```python
dt['cost'] = \
dt['COST (£) charged to Wellcome (inc VAT when charged)'].apply\
(lambda x: float(str(x).translate ({ord(c): " " for c in "£$"})))
```

```python
dt.groupby('Journal title lower')['Article title'].count().sort_values(ascending = False).head().index
```

```python
dt[dt['Journal title lower'].isin(['plos one', 'journal of biological chemistry', 'neuroimage',
                                   'plos pathogens', 'plos genetics'])].groupby('Journal title lower')['cost'].mean()
```
