```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("economicdata2017-2017.csv", index_col = 'Unnamed: 1', skiprows = 3)
columns = list(df.loc['Year'])
df.columns = columns
df = df.iloc[1:,1:]
df
```

```python
df['Economic Freedom Summary Index'] = pd.to_numeric(df['Economic Freedom Summary Index'])
df['Controls of the movement of capital and people'] = pd.to_numeric(df['Controls of the movement of capital and people'])
df['Starting a business'] = pd.to_numeric(df['Starting a business'])   
df['Business regulations'] = pd.to_numeric(df['Business regulations']) 
df.describe(include='all')
```

```python
## plots 
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(2, 2, 1)
plt.hist(df['Economic Freedom Summary Index'], bins = 10, density = True, alpha=.5)
plt.title('Economic Freedom Summary Index')

plt.subplot(2, 2, 2)
plt.hist(df['Controls of the movement of capital and people'], bins = 10, density = True, alpha=.5)
plt.title('Controls of the movement of capital and people')

plt.subplot(2, 2, 3)
plt.boxplot(df['Economic Freedom Summary Index'])

plt.subplot(2, 2, 4)
plt.boxplot(df['Controls of the movement of capital and people'])
plt.show()

# Checking the distribution of the economic freedom index & controls of movement of capotal and people, 
# most of the countries having the index lower than average. 
```

```python
df.plot(kind='scatter', x= 'Economic Freedom Summary Index',y= 'Controls of the movement of capital and people')
# Checking the realtionship between economic freedom and control of movement of capital and people,
# thought the relationship will be inverse but they aren't. 

df.plot(kind='scatter', x= 'Starting a business',y= 'Business regulations')
# same with starting a business and business regulations. Although I assumed the less regulations the more easily 
# to start a business, the points show the opposite.
plt.show()
```
